```python
import random

numbers = [2000, -500]
weights = [0.6, 0.4]

result = random.choices(numbers, weights=weights,k=1)

print(result)
print(result[0])
```

```python
sum=0
for i in range(10000000):
  result = random.choices(numbers, weights=weights, k=1)[0]
  sum+=result
```

```python
print(sum/10000000)
```
