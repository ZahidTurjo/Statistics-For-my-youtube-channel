### PMF

```python
import pandas as pd
import numpy as np
import random
```

```python
l=[]
for i in range(100000):
  a=random.randint(1,6)
  l.append(a)
```

```python
l[:20]
```

```python
s=pd.Series(l).value_counts()/100000
```

```python
s=s.sort_index()
s
```

```python
s.plot(kind='bar')
```

**CDF of PMF**

```python
np.cumsum(s).plot(kind='bar')
```

**dice roll 2 times**

```python
l=[]
for i in range(100000):
  a=random.randint(1,6)
  b=random.randint(1,6)
  l.append(a+b)
```

```python
s=pd.Series(l).value_counts()/100000
s=s.sort_index()
s.plot(kind='bar')
```

```python
np.cumsum(s).plot(kind='bar')
```

### PDF

```python
from numpy.random import normal
Normal_data = normal(loc=50, scale=5, size=1000)
```

```python
import seaborn as sns
sns.histplot(Normal_data)
```

```python
sns.kdeplot(Normal_data, shade=True)
```

**PDF and CDF**

```python
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the PDF using seaborn
sns.kdeplot(Normal_data, ax=ax1, color='b', label='PDF')

# Calculate the KDE for CDF
kde = gaussian_kde(Normal_data)
x = np.linspace(min(Normal_data), max(Normal_data), 1000)
cdf = np.cumsum(kde(x)) * (x[1] - x[0])

# Create a twin axis to plot the CDF
ax2 = ax1.twinx()
ax2.plot(x, cdf, color='r', label='CDF')

# Customize the plot
ax1.set_xlabel('Value')
ax1.set_ylabel('Density (PDF)', color='b')
ax2.set_ylabel('Cumulative Probability (CDF)', color='r')
plt.title('PDF and CDF of Sample Data')

# Show legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
